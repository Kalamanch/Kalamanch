```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("flipkart_com-ecommerce_sample.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.drop(['uniq_id', 'crawl_timestamp', 'product_url', 'pid', 'image', 'brand', 'is_FK_Advantage_product', 'product_rating', 'overall_rating', 'product_specifications'], inplace=True, axis=1)
```

```python
data
```

```python
main_category = []
sub_category = []
particular_category = []
for i in range(data.shape[0]):
    category_list = data['product_category_tree'][i].split(' >> ')
    if len(category_list)>=3:
        main_category.append(category_list[0].strip('[').strip(']').strip('"'))
        sub_category.append(category_list[1].strip('[').strip(']').strip('"'))
        particular_category.append(category_list[2].strip('[').strip(']').strip('"'))
    elif len(category_list) == 2:
        main_category.append(category_list[0].strip('[').strip(']').strip('"'))
        sub_category.append(category_list[1].strip('[').strip(']').strip('"'))
        particular_category.append(None)
    elif len(category_list) == 1:
        main_category.append(category_list[0].strip('[').strip(']').strip('"'))
        sub_category.append(None)
        particular_category.append(None)
    else:
        main_category.append(None)
        sub_category.append(None)
        particular_category.append(None)
data['main_category'] = main_category
data['sub_category'] = sub_category
data['particular_category'] = particular_category
```

```python
data.drop(['product_category_tree', 'retail_price', 'discounted_price'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.drop_duplicates(inplace=True, subset="product_name")
```

```python
data = data.reset_index()
```

```python
data.isna().sum()
```

```python
data = data.fillna('')
```

```python
tfv = TfidfVectorizer(max_features=None,
                     strip_accents='unicode',
                     analyzer='word',
                     min_df=10,
                     token_pattern=r'\w{1,}',
                     ngram_range=(1,3),
                     stop_words='english')
```

```python
tfv_matrix = tfv.fit_transform(data['description'])
```

```python
tfv_matrix
```

```python
indices = pd.Series(data.index,index=data['product_name']).drop_duplicates()
```

```python
cos = cosine_similarity(tfv_matrix,tfv_matrix)
```

```python
def product_recommendation(title, cos=cos):
    indx = indices[title]
    cos_scores = list(enumerate(cos[indx]))
    cos_scores = sorted(cos_scores, key=lambda x: x[1], reverse=True)
    cos_scores = cos_scores[1:11]
    product_indices = [i[0] for i in cos_scores]
    return data['product_name'].iloc[product_indices]
```

```python
n=input("Enter the name of the product: ")
print("\nTop Recommended products are: \n")
print(product_recommendation(n).unique())
```
